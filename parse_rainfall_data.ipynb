```python
import pandas as pd
```

## load data into notebook

```python
raw_lines = open('rainfall_data.txt', 'r'). readlines()
raw_lines
```

## parse data into a simple table

```python
lines = [raw_line.split(' ') for raw_line in raw_lines if 'Average' not in raw_line]

datapoints = []

for i, line in enumerate(lines):
    
    if i == 0:
        # parse header row
        header = lines[i]

    elif len(line) == 1:
        # parse district name
        district = line[0].strip()

    else:
        # construct data point
        for j in range(1, len(line)):
            datapoint = {
                'year': line[0],
                'rainfall_mm': line[j].strip(),
                'period': header[j].strip(),
                'district': district
            }

            datapoints.append(datapoint)


rainfall = pd.DataFrame(datapoints)        
```

## save data as a csv file

```python
rainfall.to_csv('rainfall_dataset.csv')
```
